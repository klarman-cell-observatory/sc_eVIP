```python
positions_d={}
clusters=['black','#138D75','gold','#BB8FCE','red']
for cl in clusters:
    print(cl,anno.loc[anno['cluster_color']==cl,'Position'])
    positions=list(set(anno.loc[anno['cluster_color']==cl,'Position']))
    positions.sort()
    print(cl,positions)
    positions_d[cl]=positions
```
